```python
# ==========================================================
# AI BANKING FRAUD DETECTION & RISK INTELLIGENCE SYSTEM
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

np.random.seed(42)

# ----------------------------------------------------------
# 1. GENERATE SYNTHETIC TRANSACTION DATA
# ----------------------------------------------------------

def generate_transaction_data(n=3000):
    transaction_amount = np.random.exponential(scale=2000, size=n)
    transaction_time = np.random.randint(0, 24, n)
    account_age_days = np.random.randint(10, 2000, n)
    location_change = np.random.binomial(1, 0.1, n)
    rapid_transactions = np.random.binomial(1, 0.05, n)

    fraud = (
        (transaction_amount > 8000) |
        (location_change == 1) |
        (rapid_transactions == 1)
    ).astype(int)

    return pd.DataFrame({
        "Amount": transaction_amount,
        "Hour": transaction_time,
        "AccountAgeDays": account_age_days,
        "LocationChange": location_change,
        "RapidTransactions": rapid_transactions,
        "Fraud": fraud
    })

df = generate_transaction_data()

print("Dataset Preview:")
print(df.head())

# ----------------------------------------------------------
# 2. DATA VISUALIZATION
# ----------------------------------------------------------

plt.figure(figsize=(6,5))
sns.countplot(x="Fraud", data=df)
plt.title("Fraud vs Legit Transactions")
plt.show()

# ----------------------------------------------------------
# 3. MODEL TRAINING
# ----------------------------------------------------------

X = df.drop("Fraud", axis=1)
y = df["Fraud"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=150)
model.fit(X_train, y_train)

# ----------------------------------------------------------
# 4. EVALUATION
# ----------------------------------------------------------

predictions = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n")
print(classification_report(y_test, predictions))

cm = confusion_matrix(y_test, predictions)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Reds")
plt.title("Fraud Detection Confusion Matrix")
plt.show()

# ----------------------------------------------------------
# 5. FEATURE IMPORTANCE
# ----------------------------------------------------------

importance = pd.Series(model.feature_importances_, index=X.columns)

plt.figure(figsize=(8,5))
importance.sort_values().plot(kind="barh")
plt.title("Feature Importance")
plt.show()

# ----------------------------------------------------------
# 6. RISK SCORING SYSTEM
# ----------------------------------------------------------

print("\n------ LIVE FRAUD RISK ANALYZER ------")

amount = float(input("Transaction Amount: "))
hour = int(input("Transaction Hour (0-23): "))
account_age = int(input("Account Age (days): "))
location_change = int(input("Location Change? (0/1): "))
rapid = int(input("Rapid Transactions? (0/1): "))

user_input = np.array([[amount, hour, account_age, location_change, rapid]])

prediction = model.predict(user_input)[0]
probability = model.predict_proba(user_input)[0][1]

if prediction == 1:
    print("\n⚠ High Fraud Risk Detected")
else:
    print("\n✅ Transaction Appears Legitimate")

print("Fraud Probability Score:", round(probability * 100, 2), "%")
```
